```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# on importe les informations tirées de yahoo finance sur HCP
hcp = pd.read_csv("/Users/edouardcuny/Downloads/HCP.csv")
hcp_open = hcp.iloc[:100,0:2]
```

# bollinger bands

```python
# bollinger bands
# on prend la rolling mean et on l'encadre entre 2 rolling standard deviations
# certains prennent des décisions d'investissement avec ça

ax = hcp_open.plot()

rm = hcp_open.rolling(window=20).mean()
rstd = hcp_open.rolling(window=20).std()

upper_boundary = rm.copy()
upper_boundary.iloc[:,1] += 2*rstd.iloc[:,1]
lower_boundary = rm.copy()
lower_boundary.iloc[:,1] -= 2*rstd.iloc[:,1]

rm.plot(ax=ax)
upper_boundary.plot(ax=ax)
lower_boundary.plot(ax=ax)
```

```python
plt.show()
```

# daily returns

```python
# calcul des daily returns
dr = hcp.iloc[:,1]/hcp.iloc[:,1].shift(1)-1
```

```python
dr.plot()
plt.show()
```
